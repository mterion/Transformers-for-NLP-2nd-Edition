#Summarizing documents with T5-large

```

def summarize(text,ml):
  preprocess_text = text.strip().replace("\n","") # Strip text from \n
  t5_prepared_Text = "summarize: "+preprocess_text # Add the prefix necessary for T5
  print ("Preprocessed and prepared text (look at the prefix added): \n", t5_prepared_Text)

  tokenized_text = tokenizer.encode(t5_prepared_Text, return_tensors="pt").to(device) # I saw that device was to to CPU

  # summmarize 
  summary_ids = model.generate(tokenized_text,
                                      num_beams=4, # determines the number of beams used during beam search decoding. 
                                                      # Beam search is a technique used in sequence generation tasks to explore multiple possible output sequences. 
                                                      # Setting a higher value for num_beams increases the diversity of generated sequences but also increases computation time.
                                      no_repeat_ngram_size=2, # preventing the model from generating n-grams (sequences of n tokens) that have appeared in the 
                                                      # previously generated output. Setting it to a higher value, such as 2, ensures that no 2-gram repeats 
                                                      # in the output.
                                      min_length=30, # Min length of generated output
                                      max_length=ml, # Max length of generated output
                                      early_stopping=True) # When set to True, the generation process will stop if the model predicts an end-of-sequence token or if min_length has been reached.

  output = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
  return output
```

##A general topic sample

```
text="""
The United States Declaration of Independence was the first Etext
released by Project Gutenberg, early in 1971.  The title was stored
in an emailed instruction set which required a tape or diskpack be
hand mounted for retrieval.  The diskpack was the size of a large
cake in a cake carrier, cost $1500, and contained 5 megabytes, of
which this file took 1-2%.  Two tape backups were kept plus one on
paper tape.  The 10,000 files we hope to have online by the end of
2001 should take about 1-2% of a comparably priced drive in 2001.
"""
print("Number of characters:",len(text))
summary=summarize(text,50)
print ("\n\nSummarized text: \n",summary)
```
